```python
from pyspark.sql.functions import *
combined_df=spark.read.format('delta').load('abfss://catalog@storageaccounthealthetl.dfs.core.windows.net/data/silver/device_patient_vital_streams_merge_df')
patient_360_view_df=combined_df.groupBy('patient_id').agg(count('reading_id').alias('total_readings'),countDistinct('device_id').alias('total_devices'),avg('heart_rate').alias('avg_heart_rate'),avg('bp_systolic').alias('avg_bp_sysyolic'),avg('bp_diastolic').alias('avg_bp_diastolic'),avg('temperature').alias('avg_temperature'),avg('spo2').alias('avg_spo2'),max('heart_rate').alias('max_heart_rate'),max('bp_systolic').alias('max_bp_sysyolic'),max('bp_diastolic').alias('max_bp_diastolic'),max('temperature').alias('max_temperature'),max('spo2').alias('max_spo2'),min('heart_rate').alias('min_heart_rate'),min('bp_systolic').alias('min_bp_sysyolic'),min('bp_diastolic').alias('min_bp_diastolic'),min('temperature').alias('min_temperature'),min('spo2').alias('min_spo2'))
patient_360_view_df=patient_360_view_df.withColumn('record_created_time',current_timestamp())
display(patient_360_view_df)
```
